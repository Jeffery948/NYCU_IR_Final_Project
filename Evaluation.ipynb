# LLM as Judge

```python
import csv
import json
import time
import os
from google import genai
from google.genai import types
from google.colab import userdata

# 請填入您的 API Key 或設定環境變數
# os.environ["GOOGLE_API_KEY"] = "YOUR_API_KEY"

# 您的 Ground Truth Paper Title List (請在此補全您的完整清單)
PAPER_TITLES = paper_title

class ReferenceJudge:
    def __init__(self, model_name: str = "gemini-2.5-flash"):
        self.client = genai.Client(api_key=userdata.get('Gemini'))
        self.model_name = model_name

    def verify_reference(self, model_output: str, target_title: str) -> bool:
        """
        使用 Gemini 檢查 model_output 中是否引用了 target_title
        """
        prompt = (
            "You are a strict citation verifier.\n"
            "I will provide a 'Target Paper Title' (Ground Truth) and a 'Model Output'.\n"
            "Your task is to check if the 'Model Output' correctly cites the 'Target Paper Title'.\n\n"

            "*** TARGET PAPER TITLE ***\n"
            f"{target_title}\n\n"

            "*** MODEL OUTPUT ***\n"
            f"{model_output}\n\n"

            "*** INSTRUCTIONS ***\n"
            "1. Look for a 'References' or bibliography section at the end of the Model Output.\n"
            "2. Check if the 'Target Paper Title' appears in that list.\n"
            "3. Allow for minor differences (e.g., casing, punctuation, or slight abbreviation), but the core meaning and main title words must match.\n"
            "4. If the title is present, return true. If not, return false.\n\n"

            "*** OUTPUT FORMAT ***\n"
            "Respond strictly with a JSON object:\n"
            "{\"is_reference_correct\": boolean}"
        )

        max_retries = 3
        base_wait = 5  # API 錯誤等待時間

        for attempt in range(max_retries + 1):
            try:
                response = self.client.models.generate_content(
                    model=self.model_name,
                    contents=prompt,
                    config=types.GenerateContentConfig(
                        temperature=0.0,
                        response_mime_type="application/json"
                    )
                )
                result = json.loads(response.text)
                return result.get("is_reference_correct", False)

            except Exception as e:
                if attempt < max_retries:
                    time.sleep(base_wait * (2 ** attempt))
                else:
                    print(f"  [Ref Judge Failed] Error: {e}")
                    return False

    def run_update(self, input_csv: str, output_csv: str, ground_truth_titles: list):
        if not os.path.exists(input_csv):
            print(f"Error: Input CSV '{input_csv}' not found.")
            return

        rows = []
        with open(input_csv, 'r', encoding='utf-8') as f:
            reader = csv.DictReader(f)
            rows = list(reader)
            fieldnames = reader.fieldnames

        if len(rows) > len(ground_truth_titles):
            print(f"Warning: CSV has {len(rows)} rows but Title List only has {len(ground_truth_titles)} items.")
            print("Processing will stop when titles run out.")

        print(f"Updating references for {min(len(rows), len(ground_truth_titles))} rows...")

        # 開啟輸出檔案
        with open(output_csv, 'w', newline='', encoding='utf-8') as f_out:
            writer = csv.DictWriter(f_out, fieldnames=fieldnames)
            writer.writeheader()

            for i, row in enumerate(rows):
                if i >= len(ground_truth_titles):
                    print(f"Stopping at row {i+1}: No more ground truth titles provided.")
                    # 把剩下的舊資料直接寫回去，或者選擇結束
                    writer.writerow(row)
                    continue

                target_title = ground_truth_titles[i]
                full_output = row.get("full_output", "")

                # 如果 full_output 是空的或錯誤標記，直接設為 False
                if not full_output: #or full_output.startswith("["):
                    is_correct = False
                    print(f"Row {i+1}: No valid output content. Ref Correct: False")
                else:
                    # 呼叫 Gemini 檢查
                    print(f"Row {i+1} Checking... ", end="", flush=True)
                    is_correct = self.verify_reference(full_output, target_title)
                    print(f"Ref Correct: {is_correct}")

                # 更新欄位 (注意：CSV 讀進來是字串，寫回也最好是字串)
                row["is_reference_correct"] = str(is_correct)

                writer.writerow(row)
                f_out.flush() # 確保資料寫入

        print(f"\nUpdate complete. Saved to {output_csv}")

# --- 主程式 ---
if __name__ == "__main__":
    INPUT_CSV = "vanilla.csv"      # 來源 CSV
    OUTPUT_CSV = "vanilla_new.csv"    # 輸出的新 CSV

    # 檢查 Title List 長度是否足夠，這裡只用範例資料
    # 實際使用時請確保 PAPER_TITLES 包含了完整的 50 題標題
    if len(PAPER_TITLES) < 10:
        print("Warning: PAPER_TITLES list looks very short. Please fill it in.")

    judge = ReferenceJudge()
    judge.run_update(INPUT_CSV, OUTPUT_CSV, PAPER_TITLES)
```

```python
import csv
import json
import time
import os
from google import genai
from google.genai import types
from google.colab import userdata
import sys

csv.field_size_limit(sys.maxsize)

# 請填入您的 API Key 或設定環境變數
# os.environ["GOOGLE_API_KEY"] = "YOUR_API_KEY"

# 您的 Ground Truth Paper Title List
# 假設這個 list 的順序與 litqa-v0.jsonl (扣除 canary 行後) 的順序是完全一致的
PAPER_TITLES = paper_title

class ReferenceJudge:
    def __init__(self, model_name: str = "gemini-2.5-flash"):
        # 建議使用 2.0-flash 或 pro，2.5 尚未正式普及至所有環境，若報錯請改回 1.5 或 2.0
        self.client = genai.Client(api_key=userdata.get('Gemini'))
        self.model_name = model_name

    def verify_reference(self, model_output: str, target_title: str) -> bool:
        """
        使用 Gemini 檢查 model_output 中是否引用了 target_title
        """
        prompt = (
            "You are a strict citation verifier.\n"
            "I will provide a 'Target Paper Title' (Ground Truth) and a 'Model Output'.\n"
            "Your task is to check if the 'Model Output' correctly cites the 'Target Paper Title'.\n\n"

            "*** TARGET PAPER TITLE ***\n"
            f"{target_title}\n\n"

            "*** MODEL OUTPUT ***\n"
            f"{model_output}\n\n"

            "*** INSTRUCTIONS ***\n"
            "1. Look for a 'References' or bibliography section at the end of the Model Output.\n"
            "2. Check if the 'Target Paper Title' appears in that list.\n"
            "3. Allow for minor differences (e.g., casing, punctuation, or slight abbreviation), but the core meaning and main title words must match.\n"
            "4. If the title is present, return true. If not, return false.\n\n"

            "*** OUTPUT FORMAT ***\n"
            "Respond strictly with a JSON object:\n"
            "{\"is_reference_correct\": boolean}"
        )

        max_retries = 3
        base_wait = 5

        for attempt in range(max_retries + 1):
            try:
                response = self.client.models.generate_content(
                    model=self.model_name,
                    contents=prompt,
                    config=types.GenerateContentConfig(
                        temperature=0.0,
                        response_mime_type="application/json"
                    )
                )
                result = json.loads(response.text)
                return result.get("is_reference_correct", False)

            except Exception as e:
                if attempt < max_retries:
                    time.sleep(base_wait * (2 ** attempt))
                else:
                    print(f"  [Ref Judge Failed] Error: {e}")
                    return False

    def build_id_map(self, jsonl_path: str, titles_list: list) -> dict:
        """
        讀取原始 JSONL，建立 {id: paper_title} 的對照表。
        假設 JSONL 的順序與 titles_list 的順序是一致的。
        """
        id_map = {}
        valid_ids = []

        print(f"Building ID map from {jsonl_path}...")

        with open(jsonl_path, 'r', encoding='utf-8') as f:
            for line in f:
                line = line.strip()
                if not line: continue
                try:
                    data = json.loads(line)
                    if "canary" in data: continue # 跳過 canary
                    valid_ids.append(data['id'])
                except:
                    continue

        # 檢查數量是否匹配
        if len(valid_ids) != len(titles_list):
            print(f"Warning: JSONL has {len(valid_ids)} questions but Title List has {len(titles_list)} items.")
            print("Mapping will process based on the shorter length.")

        # 建立 Map
        for i, q_id in enumerate(valid_ids):
            if i < len(titles_list):
                id_map[q_id] = titles_list[i]

        print(f"ID Map built. Total mapped items: {len(id_map)}")
        return id_map

    def run_update(self, input_csv: str, output_csv: str, jsonl_path: str, ground_truth_titles: list):
        if not os.path.exists(input_csv):
            print(f"Error: Input CSV '{input_csv}' not found.")
            return

        # 1. 建立 ID -> Title 的對照表
        ground_truth_map = self.build_id_map(jsonl_path, ground_truth_titles)

        # 2. 讀取 CSV
        rows = []
        with open(input_csv, 'r', encoding='utf-8') as f:
            reader = csv.DictReader(f)
            rows = list(reader)
            fieldnames = reader.fieldnames

        print(f"Processing {len(rows)} rows from CSV...")

        # 3. 開啟輸出檔案並逐行處理
        with open(output_csv, 'w', newline='', encoding='utf-8') as f_out:
            writer = csv.DictWriter(f_out, fieldnames=fieldnames)
            writer.writeheader()

            for i, row in enumerate(rows):
                curr_id = row.get("id")
                full_output = row.get("full_output", "")

                # 根據 ID 查找正確的 Title
                target_title = ground_truth_map.get(curr_id)

                if not target_title:
                    print(f"Row {i+1} (ID: {curr_id}): No matching ground truth title found. Skipping judge.")
                    row["is_reference_correct"] = "False" # 或標記為 Error
                    writer.writerow(row)
                    continue

                # 如果 full_output 是空的，直接判錯
                if not full_output:
                    is_correct = False
                    print(f"Row {i+1} (ID: {curr_id}): No output content. Ref: False")
                else:
                    # 呼叫 Gemini 檢查
                    print(f"Row {i+1} (ID: {curr_id}) Checking against '{target_title[:30]}...' -> ", end="", flush=True)
                    is_correct = self.verify_reference(full_output, target_title)
                    print(f"{is_correct}")

                # 更新欄位
                row["is_reference_correct"] = str(is_correct)

                writer.writerow(row)
                f_out.flush()

        print(f"\nUpdate complete. Saved to {output_csv}")

# --- 主程式 ---
if __name__ == "__main__":
    INPUT_CSV = "reflectrag_eval.csv"      # 來源 CSV (亂序沒關係)
    OUTPUT_CSV = "reflectrag_eval_new.csv" # 輸出的新 CSV
    JSONL_FILE = "litqa-v0.jsonl"          # 原始題目檔 (用來對齊 ID)

    if len(PAPER_TITLES) < 10:
        print("Warning: PAPER_TITLES list looks very short. Please check.")

    judge = ReferenceJudge()

    # 傳入 JSONL 路徑以建立對照表
    judge.run_update(INPUT_CSV, OUTPUT_CSV, JSONL_FILE, PAPER_TITLES)
```
